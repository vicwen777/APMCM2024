```python
from scipy.stats import pearsonr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from sklearn.preprocessing import MinMaxScaler
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.stats.diagnostic import acorr_ljungbox
import warnings
warnings.filterwarnings("ignore")
# Set the font to SimHei (bold) for Chinese support
plt.rcParams['font.sans-serif'] = ['SimHei']  # Chinese in bold
plt.rcParams['axes.unicode_minus'] = False    # Resolve the minus sign

def correlation_matrix_with_p_values(X):
    cols = X.columns
    # Initialize the display matrix to store the formatted (p-value) correlation coefficient
    display_matrix = np.empty((len(cols), len(cols)), dtype=object)
    display_matrix_values = np.empty((len(cols), len(cols)), dtype=object)

    # Calculate correlation coefficients and p-values
    for i in range(len(cols)):
        for j in range(len(cols)):
            # Calculate correlation coefficients and p-values
            corr, pval = pearsonr(X[cols[i]].dropna(), X[cols[j]].dropna())
            # Formatted as (p-value) correlation coefficient
            display_matrix[i, j] = f"{corr:.3f}"
            display_matrix_values[i, j] = round(corr,3)

    # Convert matrices to DataFrames
    display_df = pd.DataFrame(display_matrix, index=cols, columns=cols)
    display_df_values = pd.DataFrame(display_matrix_values, index=cols, columns=cols)

    # Draw heat maps (without numerical displays, as the data displayed is formatted by us)
    plt.figure(figsize=(15, 12))
    sns.heatmap(display_df_values.astype('float'),fmt='', cmap='coolwarm', cbar=True)
    plt.title('Correlation Between Indicators',fontproperties=latex)
    plt.xticks(fontproperties=latex)
    plt.yticks(fontproperties=latex)
    plt.savefig(f'问题4\\各指标之间的相关性.png', bbox_inches='tight')
    plt.show()
# Use functions


X = df
correlation_matrix_with_p_values(X)
```

```python
import pandas as pd
from datetime import datetime
import seaborn as sns 
sns.set(font= "simhei",style="whitegrid",font_scale=1.6)
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['axes.unicode_minus']=False # Solve the problem of displaying negative signs for coordinate axes
## Ignore the reminders 
import warnings
warnings.filterwarnings("ignore")
import scipy.stats as stats
from statsmodels.regression.linear_model import OLS
import statsmodels.api as sm

def function(data,column,name):
    # Regression equation construction
    slope, intercept, r_value, p_value, std_err = stats.linregress(data[column],data['China Export Value'])
    #Drawing
    plt.figure(figsize=(10,6))
    sns.regplot(x=column, y='China Export Value', data=data,line_kws={'label':"y={0:.3f}x+{1:.3f}\nR2={2:.3f}".format(slope,intercept,r_value**2)},color='r')
    plt.title(f'Scatterplot of China Export Value vs {name}', fontsize=18,fontproperties=latex)
    plt.tick_params(labelsize=14)
    plt.xlabel(f'{name}', fontsize=16,fontproperties=latex)
    plt.ylabel('China Export Value', fontsize=16,fontproperties=latex)
    plt.legend(fontsize='17',prop=latex)
    plt.legend(loc='upper right',fontsize=20)
    plt.savefig(f'问题4\\中国出口额{name}散点回归图.png',dpi=500) #SavePicture
    plt.show()

columns = ['China Food Production', 'USA Food', 'Western Europe Food', 'Japan Food','China Food','Western Europe Market','Japan Market','USA Market','China Market','America Cat', 'America Dog', 'France Cat',
       'France Dog', 'Germany Cat', 'Germany Dog', 'EU FDI inflows (US$ million)',
       'EU Policy Rate (%)', 'EU Trade Balance (US$ million)', 'EU Total Exports (US$ million) ', 'EU Total Imports (US$ million)',
       'USA FDI inflows (US$ million)', 'USA Policy Rate (%)', 'USA Trade Balance (US$ million)', 'USA Total Exports (US$ million) ',
       'USA Total Imports (US$ million)']

for c in columns:
    function(df,c,c)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler



# Standardized data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df[columns])

# PCA downscaling
pca = PCA()
pca_result = pca.fit_transform(data_scaled)

# Cumulative explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_.cumsum()

# Create a chart
# 1. Proportion of variance explained by each principal component
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, alpha=0.7, align='center')
plt.step(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, where='mid', label='Cumulative Variance')
plt.title('Variancecontribution Rate of Principal Components',fontproperties=latex)
plt.xlabel('Principal Components',fontproperties=latex)
plt.ylabel('Variance Accounted For',fontproperties=latex)
plt.legend(loc='best',prop=latex)
plt.savefig(f'问题4\\累积方差贡献.png',dpi=500) #SavePicture

plt.grid()

# 2. Projection of the data into the space of the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1], s=100, edgecolors='k')
plt.title('Factor Score',fontproperties=latex)
plt.xlabel('Principal Component1',fontproperties=latex)
plt.ylabel('Principal Component2',fontproperties=latex)
plt.savefig(f'问题4\\主成分因子得分.png',dpi=500) #SavePicture

plt.grid()

# Show charts
plt.show()

# Obtain PCA component matrix (feature weights corresponding to each principal component)
components_matrix = pd.DataFrame(pca.components_, 
                                 columns=[f"Feature_{i+1}" for i in range(25)],
                                 index=[f"Principal_Component_{i+1}" for i in range(pca.components_.shape[0])])

# Find the feature that contributes the most to each principal component (sorted by absolute value)
top_features_per_component = components_matrix.apply(lambda x: x.abs().nlargest(5).index.tolist(), axis=1)


components_matrix.to_excel('成分矩阵.xlsx')
components_matrix
```

#### As you can see, the regression does not work and we try to perform a Gaussian noise perturbation to extend the data

```python
import numpy as np
import pandas as pd

def augment_data_with_gaussian_noise(data, num_samples, noise_level=0.05):
    
    if isinstance(data, pd.DataFrame):
        data = data.values
    
    # The original sample was randomly selected as the basis
    sampled_indices = np.random.choice(data.shape[0], size=num_samples, replace=True)
    base_samples = data[sampled_indices]
    
    # Add Gaussian noise
    noise = np.random.normal(loc=0, scale=noise_level, size=base_samples.shape)
    augmented_samples = base_samples + noise
    
    return augmented_samples

pca_scores['y'] = y


num_new_samples = 30
noise_level = 0.01  # Noise standard deviation ratio
augmented_samples = augment_data_with_gaussian_noise(pca_scores, num_samples=num_new_samples, noise_level=noise_level)

# Merge raw and amplified data
augmented_data = pd.concat([pca_scores, pd.DataFrame(augmented_samples, columns=pca_scores.columns)], ignore_index=True)
augmented_data.to_excel('augmented_data.xlsx')
augmented_data
```
