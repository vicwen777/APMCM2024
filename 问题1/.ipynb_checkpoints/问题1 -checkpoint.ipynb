####Considering the time series forecasting model represented by autoregression, for the industry size and the number of investment and financing of pets, we use the ARIMA model to forecast, and for the number of dogs and cats, due to the shorter sequence, we use the AR model to forecast.

#### Results of Industry Scale Forecast

```python
name = []
predict = []

# 1. Read csv timing data
data = pd.Series(np.array([1000, 1500, 1700, 2500, 3500, 7000, 8000, 9000, 10000]))
data_origin = data.copy()

# 2. Use adf to test if the data is stable, if not perform first order differencing and print a plot of the data before and after differencing.
result = adfuller(data)
print("Test Statistic: %f" % result[0])
print("p-value: %f" % result[1])


    # 3. Normalize the processed data and print the normalized graph.
scaler = MinMaxScaler()
scaled_data = pd.DataFrame(scaler.fit_transform(data.values.reshape(-1, 1)))


# 4. Determine the ARIMA parameters by AIC, print the autocorrelation coefficient plots and partial autocorrelation coefficient plots for the original 
#    and post-differential data, and print the residual plots after determining the parameters.
aic_values = {}
for p in range(1,4):
    for q in range(1,4):
        try:
            model = ARIMA(scaled_data, order=(p, 1, q))
            result = model.fit()
            aic_values[(p, 1, q)] = result.aic
        except:
            continue

min_aic = min(aic_values, key=aic_values.get)
print("min aci:", min_aic)



fig, axes = plt.subplots(2, 1, figsize=(10, 8))
plot_acf(scaled_data, ax=axes[0])
plot_pacf(scaled_data, ax=axes[1],lags =3)
plt.subplots_adjust(hspace=0.3)
plt.savefig(f'问题1\\产业规模自相关偏自相关图.png',dpi=500)
plt.show()



# 5. Use the ARIMA model to make predictions, print plots of predicted and true values, and calculate model RMSE and MAPE metrics
train_size = len(scaled_data)-3
train_data, test_data = scaled_data[:train_size], scaled_data[train_size:]
model = ARIMA(train_data, order=min_aic)
result = model.fit()
predictions = result.forecast(steps=3)
predictions = scaler.inverse_transform(predictions.values.reshape(-1, 1)).reshape(1,-1)[0]

result = pd.DataFrame(predictions,columns=['产业规模'],index=['2024年','2025年','2026年'])
result.to_excel('问题1\\产业规模预测结果.xlsx')

years = ["2015", "2016", "2017", "2018", "2019", "2020", "2021", "2022", "2023"]
years = [int(i) for i in years]
years_pred = [2024,2025,2026]
industry_scale = [1000, 1500, 1700, 2500, 3500, 7000, 8000, 9000, 10000]



plt.figure(figsize=(10, 6))
plt.plot(years, industry_scale, marker="o", label="History", linestyle='-',markerfacecolor='yellow',alpha=0.8,linewidth=5,color='r')
plt.plot(years_pred, predictions, marker="o", label="Prediction", linestyle='-',markerfacecolor='yellow',alpha=0.8,linewidth=5,color='blue')
plt.xticks(rotation=0)

plt.title("Forecast Results of Industry Size", fontsize=14,fontproperties=latex)
plt.xlabel("Year", fontsize=20,fontproperties=latex)
plt.ylabel("Amount", fontsize=20,fontproperties=latex)
plt.xticks(years+years_pred,fontproperties=latex)
plt.legend(fontsize=15,prop=latex)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.savefig('问题1\\产业规模预测结果.png',dpi=500)
plt.show()
```

### Pet Cat

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import Holt

# given data
years = ["2019", "2020", "2021", "2022", "2023"]
industry_scale = [4412, 4862, 5806, 6536, 6980]

# Convert to pandas DataFrame
data = pd.DataFrame({
    'year': years,
    'industry_scale': industry_scale
})
data.set_index('year', inplace=True)


# Fitting the Holt linear trend model
# Holt method is applied to data with a trend, select the model that adds the trend ('add')
model = Holt(data['industry_scale']).fit(smoothing_level=0.8, smoothing_trend=0.2)

# Projections for the next 3 years are made
forecast_steps = 3
forecast = model.forecast(steps=forecast_steps)

# Output prediction results
forecast_years = ['2024', '2025', '2026']
forecast_df = pd.DataFrame({
    'year': forecast_years,
    'predicted_industry_scale': forecast
})
forecast_df.to_excel('问题1\\中国宠物猫走势预测结果.xlsx')
# Visualization of predicted results
plt.figure(figsize=(10, 6))
plt.plot(data['industry_scale'], marker="o", label="History", linestyle='-',markerfacecolor='yellow',alpha=0.8,linewidth=5,color='r')
plt.plot(forecast_df['year'], forecast_df['predicted_industry_scale'], marker="o", label="Prediction", linestyle='-',color='blue',markerfacecolor='yellow',alpha=0.8,linewidth=5)
plt.title("China Pet Cat Market Trend Forecast Results",fontproperties=latex)
plt.xlabel("Year",fontproperties=latex)
plt.ylabel("Amount",fontproperties=latex)
plt.legend(prop=latex)
plt.grid(True)
plt.savefig('问题1\\中国宠物猫走势预测结果.png',dpi=500)

plt.show()
```
