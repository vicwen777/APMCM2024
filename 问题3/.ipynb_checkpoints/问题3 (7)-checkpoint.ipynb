```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
from matplotlib.font_manager import FontProperties
import pandas as pd
import numpy as np
import warnings
# Ignore all warnings
warnings.filterwarnings("ignore")
# Set the font to SimHei (bold) for Chinese support
plt.rcParams['font.sans-serif'] = ['SimHei']  # Chinese in bold
plt.rcParams['axes.unicode_minus'] = False    # Resolve the minus sign
latex = FontProperties(
            family='Times New Roman',
            style='normal',
            size=12) #setting latex fonts
data = pd.read_excel('2022年全球猫狗食品进口国数据.xlsx')

# data
countries = data['国家'].values
import_values = data['进口价值（百万美元）'].values

# Plotting Horizontal Bars
plt.figure(figsize=(12, 6))
plt.barh(countries, import_values, color="blue", alpha=0.7)

# Add title and tags
plt.title("2022 Global Major Dog and Cat Food Importers by Value(Million dollar)", fontsize=14,fontproperties=latex)
plt.xlabel("Value of Imports (Million dollar)", fontsize=12,fontproperties=latex)
plt.ylabel("Country", fontsize=12,fontproperties=latex)
plt.yticks(fontproperties=latex)

# Displaying values on a bar graph
for i, value in enumerate(import_values):
    plt.text(value + 25, i, f"{value:.2f}", va="center", fontsize=10,fontproperties=latex)

# Adjusting the layout and displaying the chart
plt.tight_layout()
plt.tick_params(labelsize=13)

plt.savefig('问题3\\2022年全球主要猫狗食品进口国按价值计算.png',dpi=500)

plt.show()

```

```python
data = pd.read_excel('2022与2027年电子商务销售数据.xlsx')


# datd
countries = data['公司'].values
export_weights = data['2022年销售额（十亿美元）'].values


# data
companies = data['公司'].values
sales_2022 = data['2022年销售额（十亿美元）'].values
sales_2027 = data['2027年销售额（十亿美元）'].values

# Drawing grouped bars
bar_width = 0.35
indices = np.arange(len(companies))

plt.figure(figsize=(12, 8))
plt.barh(indices + bar_width, sales_2022, bar_width, label="2022", color="r")
plt.barh(indices, sales_2027, bar_width, label="2027", color="blue")

# Add labels, titles and legends
plt.yticks(indices + bar_width / 2, companies)
plt.yticks(fontproperties=latex)
plt.xlabel("Sales (Billions of Dollars)", fontsize=12,fontproperties=latex)
plt.title("Global Sales of Leading E-commerce Home and Pet Care Retailers, 2022 and 2027 (in Billions of Dollars)", fontsize=14,fontproperties=latex)
plt.legend(prop=latex)

# Adding values to a bar graph
for i, v in enumerate(sales_2022):
    plt.text(v + 0.5, i + bar_width, f"{v}", va="center", fontsize=10,fontproperties=latex)
for i, v in enumerate(sales_2027):
    plt.text(v + 0.5, i, f"{v}", va="center", fontsize=10,fontproperties=latex)

# Show Chart
plt.tight_layout()
plt.tick_params(labelsize=13)
plt.savefig('2022年和2027年全球领先的电子商务家居和宠物护理零售商销售额.png',dpi=500)
plt.show()
```

#### 由于这里搜集到的数据不是连续的，所以考虑使用附件3进行未来情况的预测，首先进行可视化

```python
import matplotlib.pyplot as plt


years = ["2019", "2020", "2021", "2022", "2023"]
production_values = [440.7, 727.3, 1554, 1508, 2793]  # CNY in billions
export_values = [154.1, 9.8, 12.2, 24.7, 39.6]  # USD in millions

plt.figure(figsize=(10, 6))
plt.plot(years, production_values,marker="o", linestyle='-',markerfacecolor='yellow',alpha=0.8,linewidth=5,markersize=10, label="Production Value (CNY)", color="r")
plt.plot(years, export_values ,marker="o", linestyle='-',markerfacecolor='yellow',alpha=0.8,linewidth=5,markersize=10, label="Export Value (USD)", color="blue")

plt.title("2019-2023 China Pet Food Production and Export Value(in Hundred Millions)", fontsize=14,fontproperties=latex)
plt.xlabel("Year", fontsize=12,fontproperties=latex)
plt.ylabel("Exports (in Hundred Millions)", fontsize=12,fontproperties=latex)
plt.legend(fontsize=10,prop=latex)
plt.grid(axis="y", linestyle="--", alpha=0.6)


for i, value in enumerate(production_values):
    plt.text(years[i], value + 100, f"{value}", ha='center', fontsize=10, color="b")
for i, value in enumerate(export_values):
    plt.text(years[i], value + 5, f"{value}", ha='center', fontsize=10, color="r")
plt.savefig('问题3\\2019 - 2023 年中国宠物食品产量及出口额.png',dpi=500)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import Holt

def f(years,industry_scale,name):
    # Convert to pandas DataFrame
    data = pd.DataFrame({
        'year': years,
        'industry_scale': industry_scale
    })
    data.set_index('year', inplace=True)


   # Fit Holt linear trend model
# Holt method applied to data with trend, select model with added trend ('add')
    model = Holt(data['industry_scale']).fit(smoothing_level=0.8, smoothing_trend=0.7)

    # Make projections for the next 3 years
    forecast_steps = 3
    forecast = model.forecast(steps=forecast_steps)

    # Output prediction results
    forecast_years = ['2024', '2025', '2026']
    forecast_df = pd.DataFrame({
        'year': forecast_years,
        'predicted_industry_scale': forecast
    })
    # Visualization of projected results
    plt.figure(figsize=(10, 6))
    plt.plot(data['industry_scale'], marker="o", label="History", linestyle='-',markerfacecolor='yellow',alpha=0.8,linewidth=5,color='r')
    plt.plot(forecast_df['year'], forecast_df['predicted_industry_scale'], marker="o", label="Prediction", linestyle='-',color='b',markerfacecolor='yellow',alpha=0.8,linewidth=5)
    plt.title(f"{name}",fontproperties=latex)
    plt.xlabel("Year",fontproperties=latex)
    plt.ylabel("Billion Yuan",fontproperties=latex)
    plt.legend(prop=latex)
    plt.grid(True)
    plt.savefig(f'问题3\\{name}.png',dpi=500)
    plt.show()
    return forecast_df['predicted_industry_scale'].values
```

```python
data = {
    "Year": ["2019", "2020", "2021", "2022", "2023"],
    "China's Food Production": production_values,
    "China's Export Value": export_values,
}

df = pd.DataFrame(data)
df

ls = []
for i in df.columns[1:]:
    ls.append(f(df['Year'].values.astype('str'),df[i].values,i))
result = pd.DataFrame(ls,columns=['2024', '2025', '2026'],index=df.columns[1:]).T
result.to_excel('问题3\\ 2019 - 2023 年中国宠物食品产量及出口额预测结果.xlsx')
result
```
